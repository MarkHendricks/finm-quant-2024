### A Note on String Formatting

````python
# This is one of the most useful things in python, and you will be using
# it a lot both in this class, and in the future.

# {:.2f} means to print a float with 2 decimal places.
# {:,.2f} means to print a float with 2 decimal places and commas.
# {:,.2%} means to print a float with 2 decimal places, commas, and a percent sign (multiplied by 100).
# {:,d} means to print an integer.

print("The value of pi is {:.2f}.".format(3.141592653589793))
print("Wages went up by {:,.2%}.".format(0.05))
print("The population of the US is {:,d}.".format(327200000))

# We can also use f-strings.
print("-" * 80)
print(f"The value of pi is {3.141592653589793:.2f}.")

# Finally, alignment. You will probably not use this much.
print("-" * 80)
print(f'This is left aligned: {"Hello":<10}.')
print(f'This is right aligned: {"Hello":>10}.')
print(f'This is centered: {"Hello":^10}.')

# We can also use display() to show a variable.
x = 100
display(x)
````

### 3 - Pandas

Most of your "data" will be stored in a Pandas DataFrame. In fact, most of the homework assignments (and final exam!) will be done only using the basic types (shown above) and Pandas DataFrames. Think of DataFrames as an Excel spreadsheet, but with more functionality. We have an index, which denotes the row number. This can be a label, such as a date, or just a number. We also have columns, which are the different variables. These can be strings, floats, integers, etc.

````python
# We can build DataFrames in many ways.
# If you have 2 hours: https://www.youtube.com/watch?v=pjq3QOxl9Ok
import pandas as pd

# 1. Reading an excel file.
returns = pd.read_excel(
    "../data/multi_asset_etf_data.xlsx", sheet_name="excess returns",
)

# Display the first 5 rows.
display(returns.head())

# Note how the first column is called "Date". But, we want to look up the rows by date. Thus, we need to index it by date,
# and also turn it into a datetime object.

# 1.1 Set the index to the date column.
returns["Date"] = pd.to_datetime(returns["Date"])
returns = returns.set_index("Date")

# 1.2 Load it directly with the index as the date column.
returns = pd.read_excel(
    "../data/multi_asset_etf_data.xlsx",
    sheet_name="excess returns",
    index_col=0,
    parse_dates=[0],
)

display(returns.head())
````

````python
# Resampling...
# 1. By year.
yearly_returns = returns.resample("Y").mean()

# # 2. Max return by year.
yearly_max = returns.resample("Y").max()

# # 3. Suppose we want the max return by year, but also want to know which asset it was.
yearly_max_asset = returns.idxmax(axis=1).resample("Y").first()
# yearly_max_asset
# # 4. Resample by year, then take max index AND max return.
yearly_max_asset = returns.resample("Y").agg(["idxmax", "max"])
yearly_max_asset.set_index(yearly_max_asset.index.year, inplace=True)
display(yearly_max_asset.head())
````

````python
# Subplots are also very useful.
fig_single, ax_single = plt.subplots(figsize=(7, 5))

ax_single.plot((1 + returns).cumprod() - 1)
fig.tight_layout()

# Versus a grid of suplots, with all assets except one grayed out.
fig_grid, ax_grid = plt.subplots(3, 3, figsize=(10, 10), sharex=True, sharey=True)

for i in range(3):
    for j in range(3):
        for col in returns.columns:
            ax_grid[i, j].plot(
                (1 + returns[col]).cumprod() - 1,
                linewidth=0.75,
                color="gray",
                alpha=0.3,
            )
        ax_grid[i, j].plot(
            (1 + returns.iloc[:, i * 3 + j]).cumprod() - 1, linewidth=0.75, color="blue"
        )
        ax_grid[i, j].axhline(
            0, color="red", linewidth=1, linestyle="--", label="0% Returns"
        )
        ax_grid[i, j].grid(alpha=0.5, color="gray", linestyle="--")
        ax_grid[i, j].set_title(returns.columns[i * 3 + j])
        ax_grid[i, j].tick_params(axis="x", rotation=45)

        for spine in ["top", "right"]:
            ax_grid[i, j].spines[spine].set_visible(False)


# Adjust the spacing between subplots.
fig.subplots_adjust(hspace=0.5, wspace=0.5)
plt.suptitle("Cumulative Returns")
plt.tight_layout()
````

````python
import seaborn as sns

corr_matrix = returns.corr()

# Plot the correlation matrix using seaborn's heatmap function, there are a lot of arguments here,
# so do not worry about understanding them, simply calling sns.heatmap() with the corr_matrix
# is enough to get the job done.


def plot_corr_matrix(corrs):
    return sns.heatmap(
        corrs,  # The correlation matrix
        annot=True,  # Annot means include the correlation values in the heatmap
        cmap="coolwarm",  # The color scheme to use
        vmin=-1,
        vmax=1,  # Specify the max and min values for the color scheme, otherwise they will
        # be set to the min and max values of the correlation matrix, which is not
        # particularly useful.
        linewidths=0.7,  # The width of the lines that will divide each cell, mainly an aesthetic choice,
        # but it does make the heatmap easier to read.
        annot_kws={
            "size": 10
        },  # Specify the size of the annotation text to avoid overflow into adjacent cells.
        fmt=".2f",  # Specify the format of the annotation text, in this case, we want to round to 2 decimal places.
        # See: https://kuvapcsitrd01.kutztown.edu/~schwesin/fall20/csc223/lectures/Python_String_Formatting.html
    )


fig, _ = plt.subplots(figsize=(6, 5))
ax = plot_corr_matrix(corr_matrix)
ax.xaxis.set_minor_locator(plt.NullLocator())
ax.yaxis.set_minor_locator(plt.NullLocator())

# Specify the axis to plot on, this is NOT necessary, but it is good when using multiple subplots.
ax.set_title("Correlation Matrix")  # Set the title of the plot
fig.tight_layout()  # This adjusts spacing.
````
